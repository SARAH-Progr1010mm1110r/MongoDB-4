# Index

```python
# before index
collection.find({"product_type_name": "Socks", "colour_group_name": "Pink"}).explain()["executionStats"]
```

```python
# Create compound key for the product type name and color

collection.create_index([("product_type_name", pymongo.TEXT),
                         ("colour_group_name", pymongo.TEXT)])
```

```python
# after index
collection.find({"product_type_name": "Socks", "colour_group_name": "Pink"}).explain()["executionStats"]
```

```python
# before index
collection.find({"product_type_name": "Socks", "colour_group_name": "Pink","price":72.6}).explain()["executionStats"]
```

```python
# Create compound key to improve performance for the query to retrieve product type, color, and price

collection.create_index([("product_type_name", pymongo.ASCENDING),
                         ("colour_group_name", pymongo.ASCENDING),
                         ("price",pymongo.ASCENDING)])
```

```python
# after index
collection.find({"product_type_name": "Socks", "colour_group_name": "Pink","price":72.6}).explain()["executionStats"]
```
